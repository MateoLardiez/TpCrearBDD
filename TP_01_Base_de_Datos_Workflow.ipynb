# Creación de Tablas

## Tabla Makers

```python
# MAKERS TABLE

makers = df[["make"]].drop_duplicates()
makers = makers.sort_values(by="make").reset_index().drop(columns=["index"])
makers
```

## Tabla Engine_Configuration
Se dropea la columna engine por discrepancias en los datos, más de un valor por colummna.

```python
# ENGINE CONFIGURATION TABLE

# dropeado el campo engine por discrepancias. TODO: dropear filas
engine_conf = df[["cylinders", "fuel", "transmission", "body", "doors", "drivetrain"]].drop_duplicates()
engine_conf = engine_conf.reset_index().drop(columns=["index"])
engine_conf
```

## Tabla Models

```python
# MODELS TABLE
models_partial = df.drop(columns=["price", "name", "mileage", "description", "interior_color", "exterior_color", "engine"]).drop_duplicates()

# create engine_id foreign key
def find_engine_conf(fields):
    row = engine_conf.loc[
        (engine_conf.cylinders == fields.cylinders) &
        (engine_conf.fuel == fields.fuel) &
        (engine_conf.transmission == fields.transmission) &
        (engine_conf.body == fields.body) &
        (engine_conf.doors == fields.doors) &
        (engine_conf.drivetrain == fields.drivetrain)
    ]

    return engine_conf.index.get_loc(row.index[0])

models = models_partial.copy()
models["engine_id"] = models.apply(lambda row: find_engine_conf(row), axis=1)
models = models.drop(columns=["cylinders", "fuel", "transmission", "body", "doors", "drivetrain"])
models = models.reset_index().drop(columns=["index"])

# create make foreign key
models["make"] = models.apply(lambda row: makers.loc[makers.make == row.make].index[0], axis=1)


models
```

## Tabla Classifieds

```python
# CLASSIFIEDS TABLE
c = df.drop(columns=["name", "engine"])

# create model foreign key
def find_model(fields):
    row = models_partial.loc[
        (models_partial.cylinders == fields.cylinders) &
        (models_partial.cylinders == fields.cylinders) &
        (models_partial.fuel == fields.fuel) &
        (models_partial.transmission == fields.transmission) &
        (models_partial.body == fields.body) &
        (models_partial.doors == fields.doors) &
        (models_partial.drivetrain == fields.drivetrain)
    ]
    return models_partial.index.get_loc(row.index[0])

c["model"] = c.apply(lambda row: find_model(row), axis=1)
c = c[["model", "price", "description", "interior_color", "exterior_color", "mileage"]]
c = c.reset_index().drop(columns=["index"])
c
```
